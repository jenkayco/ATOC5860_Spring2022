```python
#===============================================
# draw random values from normal distribution
#===============================================
x = np.random.normal(0,1,size = (10,))   ## small sample
y = np.random.normal(0,1,size = (3000,)) ## big sample
```

```python
plt.xlabel('value')
plt.ylabel('arbitrary')
plt.yticks([])
plt.title('values drawn from a normal distribution')
plt.xlim(-3,3)
plt.plot(x,np.arange(0,np.size(x)),'.r', markersize = 5)
plt.plot(y,np.arange(0,np.size(y)),'.k', markersize = 5)
xinc = np.arange(-4.5,5.5,.5)
for ind, val in enumerate(xinc[1:]):
    plt.plot([val - 0.25, val-0.25], [0, np.size(y)],'-', color = 'gray')
plt.show()
```

```python
## Plot Histogram Probability Distribution Function
hx = np.histogram(x,xinc)
hy = np.histogram(y,xinc)
plt.xlabel('Value')
plt.ylabel('Count')
plt.bar(hy[1][:-1],hy[0],color='k',width = .4, linewidth = 2)
plt.bar(hx[1][:-1],hx[0],color='r', width = .4, linewidth = 2)
plt.show()
```

```python
## Plot Normalized Probability Distribution Function
hx = np.histogram(x,xinc)
hy = np.histogram(y,xinc)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.bar(hx[1][:-1],hx[0].astype(float)/np.size(x),color='k', width = .4, linewidth = 2)
plt.bar(hy[1][:-1],hy[0].astype(float)/np.size(y),color='r', width = .4, linewidth = 2)
plt.show()
```

```python
## Plot Cumulative Distribution Function
hxx = np.cumsum(hx[0])
hyy = np.cumsum(hy[0])
plt.xlabel('value')
plt.ylabel('frequency')
plt.bar(hx[1][:-1],hxx.astype(float)/np.size(x),color = 'r', width = .4, linewidth = 2)
plt.bar(hy[1][:-1],hyy.astype(float)/np.size(y),color = 'k', width = .4, linewidth = 2)
plt.show()
```
