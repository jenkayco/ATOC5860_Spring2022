```python
##### ATOC7500 - variance example - Lecture #1
##### Originally coded by Prof. Barnes (CSU), updated/adapted for Jupyter Notebook/Python3 by Prof. Kay (CU)
##### Data are meteorological observations from Christman Field, Fort Collins, Colorado
##### Last updated: August 24, 2020
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd  # library for data analysis for text files (everything but netcdf files)
%matplotlib inline 
```

```python
## read in the data
filename = 'christman_2016.csv'
data=pd.read_csv(filename,sep=',')
X=data['temp_F']
```

```python
## Plot hourly temperature data
xindex = (np.arange(0,np.size(X),1))/24.;
plt.plot(xindex, X, linewidth = 0.75);
v = np.std(X);
plt.text(0.95, 0.925,'$\sigma = $' + str(np.round(v)) + '$^o$F', ha='right', transform=plt.gca().transAxes, color = 'k');
ymin, ymax = plt.ylim();
xmin, xmax = plt.xlim();
lines = plt.plot((xmin, xmax),(0, 0),(0, 0),(ymin, ymax));
plt.setp(lines,color='black', linewidth=1);
plt.ylim((ymin, ymax));
plt.xlim((xmin, xmax));
plt.ylabel('temperature (deg. F)');
plt.xlabel('day');
plt.title('Hourly temperature at Christman Field');
```

```python
### Plot hourly temperature data sampled once a day
plt.plot(xindex[::24],X[::24], linewidth = 0.74, color = 'red');
v = np.std(X[::24]);
plt.text(0.95, 0.925,'$\sigma = $' + str(np.round(v)) + '$^o$F', ha='right', transform=plt.gca().transAxes, color = 'red');
ymin, ymax = plt.ylim();
xmin, xmax = plt.xlim();
lines = plt.plot((xmin, xmax),(0, 0),(0, 0),(ymin, ymax));
plt.setp(lines,color='black', linewidth=1);
plt.ylim((ymin, ymax));
plt.xlim((xmin, xmax));
plt.ylabel('temperature (deg. F)');
plt.xlabel('days');
plt.title('Sampled once-a-day');
```

```python
### Plot hourly temperature data sampled once a month
plt.plot(xindex[::720],X[::720], linewidth = 0.74, color = 'blue');
v = np.std(X[::720]);
plt.text(0.95, 0.925,'$\sigma = $' + str(np.round(v)) + '$^o$F', ha='right', transform=plt.gca().transAxes, color = 'blue');
plt.ylabel('temperature (deg. F)');
plt.xlabel('days');
plt.title('Sampled once-a-month');
```

```python
### Plot sampled daily over one month
plt.plot(xindex[4000:4000+720:24],X[4000:4000+720:24], linewidth = 0.74, color = 'green');
v = np.std(X[4000:4000+720:24]);
plt.text(0.95, 0.925,'$\sigma = $' + str(np.round(v)) + '$^o$F', ha='right', transform=plt.gca().transAxes, color = 'green');
ymin, ymax = plt.ylim();
xmin, xmax = plt.xlim();
lines = plt.plot((xmin, xmax),(0, 0),(0, 0),(ymin, ymax));
plt.setp(lines,color='black', linewidth=1);
plt.ylim((ymin, ymax));
plt.xlim((xmin, xmax));
plt.ylabel('temperature (deg. F)');
plt.xlabel('days');
plt.title('Sampled daily over one month');
```

```python
### Sampled hourly over one day
plt.plot(xindex[4000:4000+24:1],X[4000:4000+24:1], linewidth = 0.74, color = 'magenta');
v = np.std(X[4000:4000+24:1]);
plt.text(0.95, 0.925,'$\sigma = $' + str(np.round(v)) + '$^o$F', ha='right', transform=plt.gca().transAxes, color = 'magenta');
ymin, ymax = plt.ylim();
xmin, xmax = plt.xlim();
lines = plt.plot((xmin, xmax),(0, 0),(0, 0),(ymin, ymax));
plt.setp(lines,color='black', linewidth=1);
plt.ylim((ymin, ymax));
plt.xlim((xmin, xmax));
plt.ylabel('temperature (deg. F)');
plt.xlabel('days');
plt.title('Sampled hourly over one day');
```
